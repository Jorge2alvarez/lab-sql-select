# Challenge 1

```python
books = bigquery.Client()
QUERY ='''
SELECT
    auth.au_id,
    auth.au_lname,
    auth.au_fname,
    title_n.title,
    publi.pub_name
FROM 
    `ironhack-data-analytics.publications.authors` auth
    
INNER JOIN 
    `ironhack-data-analytics.publications.titleauthor` tit_auth
ON auth.au_id = tit_auth.au_id

INNER JOIN 
    `ironhack-data-analytics.publications.titles` title_n
ON tit_auth.title_id = title_n.title_id

INNER JOIN 
    `ironhack-data-analytics.publications.publishers` publi
ON publi.pub_id = title_n.pub_id
ORDER BY
    auth.au_id,
    publi.pub_name
'''
```

```python
query_job = books.query(QUERY)
df=query_job.to_dataframe()
df.head(10)
```

# Challenge 2

```python
books = bigquery.Client()
QUERY ='''
SELECT
    auth.au_id,
    auth.au_lname,
    auth.au_fname,
    publi.pub_name,
    COUNT(title_n.title) AS count_tit
FROM 
    `ironhack-data-analytics.publications.authors` auth
    
INNER JOIN 
    `ironhack-data-analytics.publications.titleauthor` tit_auth
ON auth.au_id = tit_auth.au_id

INNER JOIN 
    `ironhack-data-analytics.publications.titles` title_n
ON tit_auth.title_id = title_n.title_id

INNER JOIN 
    `ironhack-data-analytics.publications.publishers` publi
ON publi.pub_id = title_n.pub_id
GROUP BY 
    auth.au_id,
    auth.au_lname,
    auth.au_fname,
    publi.pub_name
ORDER BY
    COUNT(title_n.title) DESC
'''
```

```python
query_job = books.query(QUERY)
df=query_job.to_dataframe()
df.head(10)
```
